# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('test.csv')
df
```

```python
df['dis'].describe()
```

```python
hightax = df.loc[df['tax'] >=666]
lowtax = df.loc[df['tax'] <= 279]
```

```python
hightax.std()
```

```python
lowtax.std()
```

lower taxed areas have a higher standard deviation in taxes and age where as higher taxed areas have a higher 
standard deviation in black people and crime

```python
closetohw = df.loc[df['rad']<=4]
farfromhw = df.loc[df['rad']>=24]
```

```python
closetohw.std()
```

```python
farfromhw.std()
```

it seems as though there is a correlation between living far from the hw and crime - this is probably because they are further from places of buisness and the people are older, so more physically and mentally vulnerable.  

```python
youngadults = df.loc[df['age']<=30]
olderadults = df.loc[df['age']>=60]
```

```python
youngadults.std()
```

```python
olderadults.std()
```

there seems to be a greater dispersion of wealth and crime amongst people over 60 than people younger than 30
you can tell this by looking at the standard deviations of the tax value and the crime rate

```python
x = df['rad'].tolist()
y = df['crim'].tolist()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel('distance from hw')
plt.ylabel('crime rate')
plt.title('distance from hw vs crime')
ax.scatter(x, y, color='blue', marker = '^')
plt.show()
```

```python
x = df['age'].tolist()
y = df['crim'].tolist()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel('age')
plt.ylabel('crime rate')
plt.title('age vs crime')
ax.scatter(x, y, color='blue', marker = '.')
plt.show()
```
